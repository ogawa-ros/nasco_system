```python
loatt_cur_file_HOT = pandas.read_csv(path_HOT+'/loatt.txt', sep=' ', header=None,names=['loatt_2l','loatt_2r','loatt_3l','loatt_3r','loatt_4l','loatt_4r','loatt_5l','loatt_5r','loatt_1l','loatt_1r'])
sis_cur_file_HOT = pandas.read_csv(path_HOT+'/sis_cur.txt', sep=' ', header=None,names=['cur_2l','cur_2r','cur_3l','cur_3r','cur_4l','cur_4r','cur_5l','cur_5r','cur_1lu','cur_1ll','cur_1ru','cur_1rl'])
sis_vol_file_HOT = pandas.read_csv(path_HOT+'/sis_vol.txt', sep=' ', header=None,names=['vol_2l','vol_2r','vol_3l','vol_3r','vol_4l','vol_4r','vol_5l','vol_5r','vol_1lu','vol_1ll','vol_1ru','vol_1rl'])
xffts_power_HOT = pandas.read_csv(path_HOT+'/xffts.txt', sep=' ', header=None,names=['power'])
loatt_cur_file_COLD = pandas.read_csv(path_COLD+'/loatt.txt', sep=' ', header=None,names=['loatt_2l','loatt_2r','loatt_3l','loatt_3r','loatt_4l','loatt_4r','loatt_5l','loatt_5r','loatt_1l','loatt_1r'])
sis_cur_file_COLD = pandas.read_csv(path_COLD+'/sis_cur.txt', sep=' ', header=None,names=['cur_2l','cur_2r','cur_3l','cur_3r','cur_4l','cur_4r','cur_5l','cur_5r','cur_1lu','cur_1ll','cur_1ru','cur_1rl'])
sis_vol_file_COLD = pandas.read_csv(path_COLD+'/sis_vol.txt', sep=' ', header=None,names=['cur_2l','cur_2r','cur_3l','cur_3r','cur_4l','cur_4r','cur_5l','cur_5r','cur_1lu','cur_1ll','cur_1ru','cur_1rl'])
xffts_power_COLD = pandas.read_csv(path_COLD+'/xffts.txt', sep=' ', header=None,names=['power'])
```

```python
d_HOT = pandas.concat([loatt_cur_file_HOT,sis_cur_file_HOT,sis_vol_file_HOT,xffts_power_HOT],axis=1)
d_COLD = pandas.concat([loatt_cur_file_COLD,sis_cur_file_COLD,sis_vol_file_COLD,xffts_power_COLD],axis=1)
```

```python
d_HOT = d_HOT[2:]
d_HOT
```

```python
bins = numpy.arange(-0.05,10.1,0.1)
bins
```

```python
def func(x):
    if x < bins[0]:return-1
    for i in range(len(bins)-1):
        if x >= bins[i] and x < bins[i+1]:return i
        continue
    return len(bins)
```

```python
d2_HOT_groupby = d_HOT.groupby('loatt_'+beam)
d2_COLD_groupby = d_COLD.groupby('loatt_'+beam)
loatt_HOT_mean =d2_HOT_groupby.mean()
loatt_COLD_mean =d2_COLD_groupby.mean()
#loatt_3l_var = d2_groupby.var()
```

```python
plt.plot(loatt_HOT_mean.index,loatt_HOT_mean['power'])
plt.plot(loatt_COLD_mean.index,loatt_COLD_mean['power'])
plt.xlabel('loatt_current(mA)')
plt.ylabel('count')
plt.show()
```

```python
Y = loatt_HOT_mean['power']/loatt_COLD_mean['power']
```

```python
Trx = (300-Y*77)/(Y-1)
```

```python
loatt_HOT_mean = loatt_HOT_mean.rename(columns = {'power':'count'})
T= pandas.concat([loatt_HOT_mean,Trx],axis = 1)
T = T.rename(columns = {'power':'Trx'})
T
```

```python
plt.plot(Trx)
plt.ylim(0,300)
plt.grid()
plt.xlabel('loatt_current(mA)')
plt.ylabel('Trx(K)')
plt.title('Trx-loatt_current_'+beam)
```

```python
xffts2loatt = {
        'xffts_power_board01': 'loatt_2l',
        'xffts_power_board02': 'loatt_2l',
        'xffts_power_board03': 'loatt_2r',
        'xffts_power_board04': 'loatt_2r',
        'xffts_power_board05': 'loatt_3l',
        'xffts_power_board06': 'loatt_3l',
        'xffts_power_board07': 'loatt_3r',
        'xffts_power_board08': 'loatt_3r',
        'xffts_power_board09': 'loatt_4l',
        'xffts_power_board10': 'loatt_4l',
        'xffts_power_board11': 'loatt_4r',
        'xffts_power_board12': 'loatt_4r',
        'xffts_power_board13': 'loatt_5l',
        'xffts_power_board14': 'loatt_5l',
        'xffts_power_board15': 'loatt_5r',
        'xffts_power_board16': 'loatt_5r',
}
# board04200GHz(CLT:9/16)
```

```python
df_ = {
    _ : pandas.concat(
        [
            pandas.DataFrame(
                loatt[xffts2loatt[_]]['data'],
                index = pandas.to_datetime(loatt[xffts2loatt[_]]['received_time'], unit='s'),
                columns = ['loatt'],
            ),
            
            pandas.DataFrame(
                xffts[_]['total_power'],
                index = pandas.to_datetime(xffts[_]['received_time'], unit='s'),
                columns = ['tp'],
            ),
        ],

        axis = 1,
    )
    for _ in xffts_topics
}

df = {_ : df_[_].resample('0.1S').median() for _ in df_}




```

```python
df_hot = {}
df_cold = {}
for _ in xffts_topics:
     df_hot[_] = df[_][(df[_].index > datetime.datetime.fromtimestamp(t1))&(df[_].index < datetime.datetime.fromtimestamp(t2))]
for _ in xffts_topics:
     df_cold[_] = df[_][(df[_].index > datetime.datetime.fromtimestamp(t2))&(df[_].index < datetime.datetime.fromtimestamp(t3))]
```

```python
fig = matplotlib.pyplot.figure(figsize=(20,10))
ax =[fig.add_subplot(4,4, i) for i in range(1,17)]
for i,(_ax,key) in enumerate(zip(ax, df)):
    _ax.plot(df_hot[key]['loatt'], df_hot[key]['tp'], '.', mfc= 'C1', mec='None')
    _ax.plot(df_cold[key]['loatt'], df_cold[key]['tp'], '.', mfc= 'C0', mec='None')
    if i //4 == 3:
        _ax.set_xlabel('current( A)')
    if i%4 == 0:
        _ax.set_ylabel('count')
    _ax.text(0.05, 0.9, key,size=15, transform=_ax.transAxes)
    _ax.grid()
fig.suptitle(mode+'/'+date+' :: power', size=20)
fig.savefig(mode+'_'+date+'_loatt_tp.png')
```

```python
c_hot = {}
c_cold = {}
for _ in xffts_topics:
    c_hot[_] = pandas.cut(df_hot[_]['loatt'], numpy.arange(0,10.1,0.1))  #(initialize_voltage, finalize_voltage+0.1, step)
    c_cold[_] = pandas.cut(df_cold[_]['loatt'], numpy.arange(0,10.1,0.1))
```

```python
df_hot_ ={}
df_cold_ ={}
for _ in xffts_topics:
    df_hot_[_] = df_hot[_].groupby(c_hot[_]).mean()
    df_cold_[_] = df_cold[_].groupby(c_cold[_]).mean()
```

```python
fig = matplotlib.pyplot.figure(figsize=(20,10))
ax =[fig.add_subplot(4,4, i) for i in range(1,17)]
for i,(_ax,key) in enumerate(zip(ax, df_hot_)):
    _ax.plot(df_hot_[key]['loatt'] ,10*numpy.log10(df_hot_[key]['tp']/df_cold_[key]['tp']),'.')
    if i //4 == 3:
        _ax.set_xlabel('loatt current(A )')
    if i%4 == 0:
        _ax.set_ylabel('R-SKY (dB)')
    _ax.text(0.05, 0.9, key,size=15, transform=_ax.transAxes)
    _ax.grid()
fig.suptitle(mode+'/'+date+' :: R-SKY(dB)', size=20)
fig.savefig(mode+'_'+date+'_loatt_rsky.png')
```

```python
fig = matplotlib.pyplot.figure(figsize=(20,10))
ax =[fig.add_subplot(4,4, i) for i in range(1,17)]
for i,(_ax,key) in enumerate(zip(ax, df_hot_)):
    Tsys = 300/(df_hot_[key]['tp']/df_cold_[key]['tp'] -1)
    _ax.plot(df_hot_[key]['loatt'] ,Tsys, '.')
    if i //4 == 3:
        _ax.set_xlabel('loatt current(mA )')
    if i%4 == 0:
        _ax.set_ylabel('Tsys (K)')
    _ax.set_ylim(0, 2000)
    _ax.text(0.05, 0.9, key,size=15, transform=_ax.transAxes)
    _ax.grid()
fig.suptitle(mode+'/'+date+' :: Tsys(K)', size=20)
fig.savefig(mode+'_'+date+'.png')
```

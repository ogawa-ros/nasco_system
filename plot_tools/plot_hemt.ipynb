```python
import numpy
import matplotlib

%matplotlib inline
import matplotlib.pyplot
matplotlib.rcParams['figure.facecolor'] = 'white'
```

```python
path = '/home/necst/data/experiments/logger/hemt_sweep/20181011_142155'
```

```python
hemt_list = ['2l', '2r', '3l', '3r','4l', '4r', '5l', '5r', '1lu', '1ll', '1ru', '1rl']
hemt = '4l'
```

```python
hemt_vg1 = numpy.loadtxt(path + '/hemt_vg1.txt')
hemt_vg2 = numpy.loadtxt(path + '/hemt_vg2.txt')
hemt_id = numpy.loadtxt(path + '/hemt_id.txt')
hemt_power = numpy.loadtxt(path + '/power.txt')
timelist = numpy.loadtxt(path + '/datatime.txt')
```

```python
vg1 = hemt_vg1[:, hemt_list.index(hemt)]
vg2 = hemt_vg2[:, hemt_list.index(hemt)]
ID = hemt_id[:, hemt_list.index(hemt)]
if hemt_list.index(hemt) % 2 == 0:
    power = hemt_power[:, 1] ###????
else:
    power = hemt_power[:, 0] ###???
```

```python
idmask = (ID <= 2.2) & (ID >= 1.8)
```

```python
ncol = 2
nrow = 1
nax = ncol * nrow
```

```python
fig = matplotlib.pyplot.figure(figsize = (ncol * 8, nrow * 8))
ax = [fig.add_subplot(nrow, ncol, i+1) for i in range(nax)]
    
ax[0].plot(timelist[idmask], power[idmask], "o")
ax[0].set_title("Power (1.9 <= id <= 2.1)")
ax[0].set_xlabel("time [s]")
ax[0].set_ylabel("Power")
ax[1].plot(timelist, power, "o")
ax[1].set_title("Power (all)")
ax[1].set_xlabel("time [s]")
ax[1].set_ylabel("Power")

    
[a.grid() for a in ax]

matplotlib.pyplot.show()
```

```python
index_max = numpy.argmax(power[idmask])

vg1_max = vg1[idmask][index_max]
vg2_max = vg2[idmask][index_max]
id_max = ID[idmask][index_max]
power_max = power[idmask][index_max]
print("vg1 : ", vg1_max, "         vg2 : ", vg2_max, "  [V]")
print("id : ", id_max, "         power : ", power_max)
index_max
```

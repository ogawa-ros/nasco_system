```python
def plot_all( saveto=None, dpi=None):
    
    sis_list = ['2l', '2r', '3l', '3r',
                      '4l', '4r', '5l', '5r', 
                      '1lu', '1ll', '1ru', '1rl']
    
    voltage = numpy.loadtxt('/home/amigos/data/experiments/logger/sisiv/20181101_085220/sis_vol.txt')
    current = numpy.loadtxt('/home/amigos/data/experiments/logger/sisiv/20181101_085220/sis_cur.txt')
    
    
    
    ncol = 4
    nrow = 3
    nax = ncol * nrow
    
    matplotlib.rcParams['savefig.dpi'] = 50
    matplotlib.rcParams['font.size'] = 24
    
    figsize = (ncol * 8, nrow * 8)
    
    fig = matplotlib.pyplot.figure(figsize=figsize)
    ax = [fig.add_subplot(nrow, ncol, i+1) for i in range(nax)]
    
    for _ax, sis, _ in zip(ax, sis_list, range(len(sis_list))):
        _ax.plot(voltage[:, _], current[:, _], '.')
        _ax.grid()
        _ax.set_xlabel('vol [mV]')
        _ax.set_ylabel('cur [uA]')
        _ax.set_title('{}'.format(sis))
        
    fig.subplots_adjust(wspace=0.4, hspace=0.4)
    fig.suptitle('\nNASCO IV MEASUREMENT on ROS', size=48)
    
#     matplotlib.pyplot.savefig('./fig/{}.png'.format(path))
        
    matplotlib.pyplot.show()
```

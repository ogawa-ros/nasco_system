```python
import numpy
import matplotlib

%matplotlib inline
import matplotlib.pyplot
matplotlib.rcParams['figure.facecolor'] = 'white'
```

```python
path_hot = '/home/amigos/data/experiments/logger/sisv_sweep/20181019_120129'
path_cold = '/home/amigos/data/experiments/logger/sisv_sweep/20181019_120129'
```

```python
power = numpy.loadtxt(path_hot + '/power.txt')
timelist = numpy.loadtxt(path_hot + '/datatime.txt')
voltagelist = numpy.loadtxt(path_hot + '/sis_vol.txt')
voltagelist2 = numpy.loadtxt(path_cold + '/sis_vol.txt')
currentlist = numpy.loadtxt(path_hot + '/sis_cur.txt')
#_currentlist = numpy.loadtxt()
power2 = numpy.loadtxt(path_cold + '/power.txt')
```

```python
ncol = 2
nrow = 1
nax = ncol * nrow
```

```python
print("AVERAGE")
avelist = [numpy.average(power[:, 0]).round(5), numpy.average(power[:, 1]).round(5)]
print(avelist, "\n")

print("VARIANCE")
varlist = [numpy.var(power[:, 0]).round(5), numpy.var(power[:, 1]).round(5)]
print(varlist, "\n")
```

```python
fig = matplotlib.pyplot.figure(figsize = (ncol * 8, nrow * 8))
ax = [fig.add_subplot(nrow, ncol, i+1) for i in range(nax)]
    
ax[0].plot(timelist, power[:, 0], "o")
#ax[0].plot()
ax[0].set_title("Power")
ax[0].set_xlabel("time [s]")
# ax[0].set_xlim(35 + 1.5393372e9,45 + 1.5393372e9)
ax[0].set_ylabel("Power")
ax[1].plot(timelist, power[:, 1], "o")
ax[1].set_title("Power")
ax[1].set_xlabel("time [s]")
ax[1].set_ylabel("Power")

    
[a.grid() for a in ax]

matplotlib.pyplot.show()
```

```python
fig = matplotlib.pyplot.figure(figsize = (ncol * 8, nrow * 8))
ax = [fig.add_subplot(nrow, ncol, i+1) for i in range(nax)]
    
ax[0].plot(voltagelist[:, 6], power[:, 0] ,'-o')
ax[0].plot(voltagelist2[:,6],power2[:,0],'-o')
ax0 = ax[0].twinx()
ax0.plot(voltagelist[:, 6], currentlist[:, 6], "-o", color="g")
#ax[0].set_xlim(6,9)
ax[0].set_title("Power")
ax[0].set_xlabel("voltage [mV]")
# ax[0].set_xlim(35 + 1.5393372e9,45 + 1.5393372e9)
ax[0].set_ylabel("Power")
ax[1].plot(voltagelist[:, 7], power[:, 1],'-o')
ax[1].plot(voltagelist2[:, 7], power2[:, 1] ,'-o')
ax1 = ax[1].twinx()
ax1.plot(voltagelist[:, 7], currentlist[:, 7], "-o", color="g")
#ax[1].set_xlim(7,9)
ax[1].set_title("Power")
ax[1].set_xlabel("voltage [mV]")
ax[1].set_ylabel("Power")

    
[a.grid() for a in ax]

matplotlib.pyplot.show()
```

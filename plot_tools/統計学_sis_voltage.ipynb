```python
import pandas
import numpy
import matplotlib.pyplot as plt
```

```python
path_HOT = '/home/amigos/data/experiments/logger/sisv_sweep/20181027_110950'
path_COLD = '/home/amigos/data/experiments/logger/sisv_sweep/20181027_111120'
```

```python
beam = '3l'

beam_list = ['2l', '2r', '3l', '3r',
                       '4l', '4r', '5l', '5r', 
                       '1lu', '1ll', '1ru', '1rl']

loatt_beam_list = ['2l', '2r', '3l', '3r',
                       '4l', '4r', '5l', '5r', 
                       '1l', '1r' ]
```

```python
loatt_cur_file_HOT = pandas.read_csv(path_HOT+'/loatt.txt', sep=' ', header=None,names=['loatt_2l','loatt_2r','loatt_3l','loatt_3r','loatt_4l','loatt_4r','loatt_5l','loatt_5r','loatt_1l','loatt_1r'])
sis_cur_file_HOT = pandas.read_csv(path_HOT+'/sis_cur.txt', sep=' ', header=None,names=['cur_2l','cur_2r','cur_3l','cur_3r','cur_4l','cur_4r','cur_5l','cur_5r','cur_1lu','cur_1ll','cur_1ru','cur_1rl'])
sis_vol_file_HOT = pandas.read_csv(path_HOT+'/sis_vol.txt', sep=' ', header=None,names=['vol_2l','vol_2r','vol_3l','vol_3r','vol_4l','vol_4r','vol_5l','vol_5r','vol_1lu','vol_1ll','vol_1ru','vol_1rl'])
xffts_power_HOT = pandas.read_csv(path_HOT+'/xffts.txt', sep=' ', header=None,names=['power'])
loatt_cur_file_COLD = pandas.read_csv(path_COLD+'/loatt.txt', sep=' ', header=None,names=['loatt_2l','loatt_2r','loatt_3l','loatt_3r','loatt_4l','loatt_4r','loatt_5l','loatt_5r','loatt_1l','loatt_1r'])
sis_cur_file_COLD = pandas.read_csv(path_COLD+'/sis_cur.txt', sep=' ', header=None,names=['cur_2l','cur_2r','cur_3l','cur_3r','cur_4l','cur_4r','cur_5l','cur_5r','cur_1lu','cur_1ll','cur_1ru','cur_1rl'])
sis_vol_file_COLD = pandas.read_csv(path_COLD+'/sis_vol.txt', sep=' ', header=None,names=['vol_2l','vol_2r','vol_3l','vol_3r','vol_4l','vol_4r','vol_5l','vol_5r','vol_1lu','vol_1ll','vol_1ru','vol_1rl'])
xffts_power_COLD = pandas.read_csv(path_COLD+'/xffts.txt', sep=' ', header=None,names=['power'])
```

```python
d_HOT = pandas.concat([loatt_cur_file_HOT,sis_cur_file_HOT,sis_vol_file_HOT,xffts_power_HOT],axis=1)
d_COLD = pandas.concat([loatt_cur_file_COLD,sis_cur_file_COLD,sis_vol_file_COLD,xffts_power_COLD],axis=1)
```

```python
plt.plot(d_HOT['vol_'+beam],d_HOT['power'],)
plt.plot(d_COLD['vol_'+beam],d_COLD['power'],'.')
```

```python
bins = numpy.arange(6.95,9.1,0.1)
range(len(bins))
bins
```

```python
def func(x):
    if x < bins[0]:return-1
    for i in range(len(bins)-1):
        if x >= bins[i] and x < bins[i+1]:return i
        continue
    return len(bins)
```

```python
d_HOT['x'] = d_HOT['vol_'+beam].apply(func)
d_COLD['x'] = d_COLD['vol_'+beam].apply(func)
d_HOT
d_COLD
```

```python
d2_HOT_groupby = d_HOT.groupby('x')
d2_COLD_groupby = d_COLD.groupby('x')
vol_HOT_mean =d2_HOT_groupby.mean()
vol_COLD_mean =d2_COLD_groupby.mean()
#loatt_3l_var = d2_groupby.var()
vol_HOT_mean
```

```python
vol_HOT_mean['vol_'+beam]
```

```python
fig, ax1 = plt.subplots()
ax1.plot(vol_COLD_mean['vol_'+beam],vol_COLD_mean['power'], label='COLD')
ax1.plot(vol_HOT_mean['vol_'+beam],vol_HOT_mean['power'],label='HOT')
ax2 = ax1.twinx()
ax2.plot(vol_HOT_mean['vol_'+beam],vol_HOT_mean['cur_'+beam] , c='green', label='sis_current')
ax2.plot()
plt.grid()
ax1.legend(loc=4)
ax2.legend()
ax1.set_xlabel('sis_vol(mV)')
ax1.set_ylabel('count')
ax2.set_ylabel('sis_current(mA)')
ax1.set_title(beam)
plt.show()
```
